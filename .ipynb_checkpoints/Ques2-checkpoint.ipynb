```python
from sklearn.cluster import KMeans
kmeans = KMeans(init='k-means++',n_clusters=4,max_iter=150,n_init=10,random_state=52)
```

```python
X = df.iloc[:,[0,1,2,3]].values
```

```python
y_kmeans = kmeans.fit_predict(X)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], 
            s = 50, c = 'pink', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], 
            s = 50, c = 'black', label = 'Iris-versicolour')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1],
            s = 50, c = 'red', label = 'Iris-virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 50, c = 'green', label = 'Centroids')

plt.legend()
```
